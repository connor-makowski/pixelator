```python
input_image = '/content/drive/MyDrive/Development/pixelator/cat.jpg'
output_image = '/content/drive/MyDrive/Development/pixelator/lego_cat.jpg'
output_data = '/content/drive/MyDrive/Development/pixelator/lego_cat.csv'

# Specify the desired output width and height in inches
width=8.5
height=11

# Specify a BGR Color Palette
palette=[[0,0,0],[255,255,255]]

# Specify the size (width or length) in inches of a 1x1 Lego
lego_size=0.314961
```

```python
from pixelator import Pixelator
from pamda import pamda
image = Pixelator(filename=input_image)
pixelated_image = image.pixelate(
    width=round(width/lego_size),
    height=round(height/lego_size),
    palette=palette
)
pixelated_image.write(filename=output_image, width=int(300*(width/height)), height=int(300*(height/width)))
pamda.write_csv(
    filename=output_data,
    data=[{'color':key, 'count':value} for key, value in pixelated_image.get_color_counts().items()]
)
```
